```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import re
import joblib
import string
```

```python
DATA_PATH = "cleaned_welfake_news_dataset.csv"  

TEXT_COL = "text" 
LABEL_COL = "class"
```

```python
df = pd.read_csv(DATA_PATH)
```

```python
# Drop rows with missing values in text or label
# Also drop rows with very short text (<=20 chars)
df = df.dropna(subset=[TEXT_COL, LABEL_COL])
df = df[df[TEXT_COL].str.len() > 20].reset_index(drop=True)
```

```python
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
df = df[['text','class']]
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'\W', ' ', text)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'<.*?>+', '', text)
    text = re.sub("[%s]" % re.escape(string.punctuation), "", text)
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'\w*\d\w*', '', text)
    return text
```

```python
df["text"]= df["text"].apply(lambda x: clean_text(str(x)) if not pd.isnull(x) else "")
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.to_csv("cleaned_welfake_news_dataset.csv", index=False)
```
